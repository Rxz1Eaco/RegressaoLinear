```python
import pandas as pd 
df = pd.read_csv("Advertising.csv")
display(df)
```

Regressão Linear: TV

```python
x = df['TV (R$ mil)']
y = df['Vendas (R$ mil)']
x_2 = df['TV (R$ mil)']**2
y_2 = df['Vendas (R$ mil)']**2
xy = x*y

dados_novo_df = {
    'X':x,
    'Y':y,
    'x^2': x_2,
    'y^2':y_2,
    'xy':xy
}
 
df_novo = pd.DataFrame(dados_novo_df)
display(df_novo)
```

Coefieciente de Correlação

```python
n = len(df_novo)
somatario_x  = df_novo['X'].sum()
somatario_y = df_novo['Y'].sum()
somatario_x_2 = df_novo['x^2'].sum()
somatario_y_2 = df_novo['y^2'].sum()
somatario_xy = df_novo['xy'].sum()


rxy = ((n * somatario_xy) - (somatario_x * somatario_y)) / (((n * somatario_x_2) - (somatario_x ** 2)) ** 0.5 * ((n * somatario_y_2) - (somatario_y ** 2)) ** 0.5)
print(rxy)
```

Coeficiente Angular

```python
b = ((n*somatario_xy) - (somatario_x)*(somatario_y))/((n*somatario_x_2)-(somatario_x**2))  
print(b)
```

```python
a = (somatario_y - b * somatario_x)/n
print(a)
```
